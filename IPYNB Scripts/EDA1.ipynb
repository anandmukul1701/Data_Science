# Data Type Conversion

```python
data.info()
```

```python
data['Month']=pd.to_numeric(data['Month'],errors='coerce')
data['Temp C']=pd.to_numeric(data['Temp C'], errors = 'coerce')
data['Weather']=data['Weather'].astype('category')
```

# Missing Values & Imputation

```python
import seaborn as sns
cols = data_cleaned3.columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. 
sns.heatmap(data_cleaned3[cols].isnull(),cmap=sns.color_palette(colours))
```

```python
data_cleaned3[data_cleaned3.isnull().any(axis=1)].head()
```

```python
data_cleaned3.isnull().sum()
```

```python
# Mean Imputation
mean = data_cleaned3['Ozone'].mean()
print(mean)
```

```python
data_cleaned3['Ozone'] = data_cleaned3['Ozone'].fillna(mean)
```

```python
data_cleaned3.isnull().sum()
```

```python
# Missing values imputation for categorical variable
# Get the object Columns
obj_columns=data_cleaned3[['Weather']]
```

```python
obj_columns.isnull().sum()
```

```python
obj_columns.mode()
```

```python
obj_columns.mode().iloc[0]
```

```python
obj_columns=obj_columns.fillna(obj_columns.mode().iloc[0])
```

```python
obj_columns.isnull().sum()
```

```python
# Join the dataset with imputed object dataset
data_cleaned4=pd.concat([data_cleaned3,obj_columns],axis=1)
```

```python
data_cleaned4.isnull().sum()
```

# Normalization of the data

```python
from numpy import set_printoptions
from sklearn.preprocessing import MinMaxScaler
```

```python
data_cleaned4.values
```

```python
array = data_cleaned4.values

scaler = MinMaxScaler(feature_range=(0,1))
rescaledX = scaler.fit_transform(array)
```

```python
# transformed data
set_printoptions(precision=2)
print(rescaledX[0:5,:])
```

```python
# Standardaize data (0 mean, 1 stdev)
from sklearn.preprocessing import StandardScaler
```

```python
array = data_cleaned4.values
scaler = StandardScaler().fit(array)
rescaledX = scaler.transform(array)

# summarize transformed data
set_printoptions(precision=2)
print(rescaledX[0:5,:])
```
