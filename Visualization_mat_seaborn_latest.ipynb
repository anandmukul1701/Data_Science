**Seaborn**

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

*Univariate :: One Variable Data Visualization*

```python
# Strip plot

sns.stripplot(y = 'tip', data = tips, jitter = True)
plt.ylabel('Tip ($)')
plt.show()
```

```python
# Grouping with stripplot()
sns.stripplot(x = 'day', y= 'tip', data = tips)
plt.ylabel('Tips($)')
plt.show()
```

```python
sns.stripplot(x = 'day', y= 'tip', data = tips, size = 4, jitter = False)
plt.ylabel('Tips($)')
plt.show()
```

```python
# Swarm plot
sns.swarmplot(x='day', y = 'tip', data = tips)
plt.ylabel('Tips ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y = 'tip', data = tips, hue = 'sex')
plt.ylabel('Tips ($)')
plt.show()
```

```python
# Changing Orientation
sns.swarmplot(x='tip', y = 'day', data = tips, hue = 'sex', orient = 'h')
plt.ylabel('Tips ($)')
plt.show()
```

# Box & Violin Plot

```python
# Boxplot
sns.boxplot(x= 'day', y = 'tip', data=tips)
plt.ylabel('Tips($)')
```

```python
# Violinplot
sns.violinplot(x= 'day', y = 'tip', data=tips)
plt.ylabel('Tips($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x= 'day', y = 'tip', data=tips)
plt.ylabel('Tips($)')

plt.subplot(1,2,2)
sns.violinplot(x= 'day', y = 'tip', data=tips)
plt.ylabel('Tips($)')
plt.show()
```

```python
## Combining plots 
sns.violinplot(x= 'day', y = 'tip', data=tips, inner = None, color = 'lightgray')
sns.stripplot(x = 'day', y= 'tip', data = tips, size = 4, jitter = True)
plt.ylabel('Tips($)')
plt.show()

```

Bivariate :: Two Variables data visualization

```python
# Joint Plot

sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
# Density Plot

sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
plt.show()
```

```python
# Pair Plot

sns.pairplot(tips)
plt.show()
```

```python
# Pair Plot

sns.pairplot(tips, hue = 'sex')
plt.show()
```
