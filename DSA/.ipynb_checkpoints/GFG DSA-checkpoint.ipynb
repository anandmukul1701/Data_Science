<p>Given an array arr[] of n integers, construct a Product Array prod[] (of the same size) such that prod[i] is equal to the product of all the elements of arr[] except arr[i].</p>
<p><strong>Note:</strong> Solve it <em>without the division operator</em> in O(n) time.</p>
<h3>Example:</h3>

<p><strong>Input:</strong> arr[] = {10, 3, 5, 6, 2}<br>
<strong>Output:</strong> prod[] = {180, 600, 360, 300, 900}<br>
3 * 5 * 6 * 2 product of other array elements except 10 is 180<br>
10 * 5 * 6 * 2 product of other array elements except 3 is 600<br>
10 * 3 * 6 * 2 product of other array elements except 5 is 360<br>
10 * 3 * 5 * 2 product of other array elements except 6 is 300<br>
10 * 3 * 6 * 5 product of other array elements except 2 is 900</p>

<p><strong>Input:</strong> arr[] = {1, 2, 3, 4, 5}<br>
<strong>Output:</strong> prod[] = {120, 60, 40, 30, 24}<br>
2 * 3 * 4 * 5 product of other array elements except 1 is 120<br>
1 * 3 * 4 * 5 product of other array elements except 2 is 60<br>
1 * 2 * 4 * 5 product of other array elements except 3 is 40<br>
1 * 2 * 3 * 5 product of other array elements except 4 is 30<br>
1 * 2 * 3 * 4 product of other array elements except 5 is 24</p>

```python
def ProdArr(arr):
    l = len(arr)
    result = []
    for i in range(l):
        prod = 1
        for j in range(l):
            if j != i:
                prod*=arr[j]
        result.append(prod)
    return result

A1 = [10, 3, 5, 6, 2]
A2 = [1, 2, 3, 4, 5]
```

<h1>Maximum Product Subarray</h1>

<strong>Explanation:</strong> The subarray is {60}</p>

```python
def MaxProductSubarray(a):
    max_product_val = float('-inf')
    max_ending_here = 1
    size = len(a)
    max_vals = []
    for i in range(0,size):
        max_ending_here = max_ending_here * a[i]
        if max_product_val < max_ending_here:
            max_product_val = max_ending_here
            max_vals.append(max_product_val)
        if max_ending_here == 0:
            max_ending_here = 1

    return max(max_vals)

a = [-1, -3, -10, 0, 60] 
```

<h2>Find all triplets with zero sum</h2>
<p><strong>Last Updated : 26 Apr, 2024</strong></p>
<p>Given an array of distinct elements. The task is to find triplets in the array whose sum is zero.</p>
<h3>Examples :</h3>
<p><strong>Input:</strong> arr[] = {0, -1, 2, -3, 1}<br>
<strong>Output:</strong> (0 -1 1), (2 -3 1)<br>
<strong>Explanation:</strong> The triplets with zero sum are 0 + -1 + 1 = 0 and 2 + -3 + 1 = 0</p>
<p><strong>Input:</strong> arr[] = {1, -2, 1, 0, 5}<br>
<strong>Output:</strong> 1 -2 1<br>
<strong>Explanation:</strong> The triplets with zero sum is 1 + -2 + 1 = 0</p>

```python
def TripletWithZeroSum(arr):
    l = len(arr)
    result = []
    for i in range(l):
        for j in range(i+1,l):
            for k in range(j+1,l):
                if (arr[i] + arr[j] + arr[k]) == 0:
                    result.append((arr[i],arr[j],arr[k]))
    return result

a1 = [0, -1, 2, -3, 1]
a2 = [1, -2, 1, 0, 5]
```

```python
print(TripletWithZeroSum(a1))
print(TripletWithZeroSum(a2))
```

<h2>Container with Most Water</h2>
<p><strong>Last Updated : 20 Jan, 2022</strong></p>
<p>Given n non-negative integers <code>a<sub>1</sub>, a<sub>2</sub>, ..., a<sub>n</sub></code> where each represents a point at coordinate <code>(i, a<sub>i</sub>)</code>, 'n' vertical lines are drawn such that the two endpoints of line i is at <code>(i, a<sub>i</sub>)</code> and <code>(i, 0)</code>. Find two lines, which together with x-axis forms a container, such that the container contains the most water. The program should return an integer which corresponds to the maximum area of water that can be contained (maximum area instead of maximum volume sounds weird but this is the 2D plane we are working with for simplicity).</p>
<p><strong>Note:</strong> You may not slant the container.</p>
<h3>Examples :</h3>
<p><strong>Input:</strong> array = [1, 5, 4, 3]<br>
<strong>Output:</strong> 6<br>
<strong>Explanation:</strong><br>
5 and 3 are distance 2 apart.<br>
So the size of the base = 2.<br>
Height of container = min(5, 3) = 3.<br>
So total area = 3 * 2 = 6</p>
<p><strong>Input:</strong> array = [3, 1, 2, 4, 5]<br>
<strong>Output:</strong> 12<br>
<strong>Explanation:</strong><br>
5 and 3 are distance 4 apart.<br>
So the size of the base = 4.<br>
Height of container = min(5, 3) = 3.<br>
So total area = 4 * 3 = 12</p>


```python
def ContainerMostWater(a):
    size = len(a)
    max_area = 0
    for i in range(0,size):
        for j in range(i,size):
            area_i_j = (j-i)*min(a[i],a[j])
            if area_i_j > max_area:
                max_area = area_i_j

    return max_area

a1 = [1, 5, 4, 3]
a2 = [3, 1, 2, 4, 5]
```

```python
print(ContainerMostWater(a1))
print(ContainerMostWater(a2))
```

```python
def TrappingRainWater(a):
    size = len(a)
    capacity = 0
    for i in range(0,size):
        # Extreme blocks will not hold any water
        if (i == 0 or i == (size - 1)):
            pass
        # Interior blocks will hold n unit of water, n = number of units possible on that block
        # An interior block whose height is equal to or more than max height block on both sides will not hold any water
        else:
            # print(i)
            max_left = max(a[:i])
            # print(max_left)
            max_right = max(a[i+1:])
            # print(max_right)
            capacity_i = max(min(max_left,max_right) - a[i], 0)
            # print(capacity_i)
            capacity = capacity+capacity_i

    return capacity

a1 = [3, 0, 1, 0, 4, 0, 2]
a2 = [3, 0, 2, 0, 4]
```

<h2>Chocolate Distribution Problem</h2>
<p><strong>Last Updated : 26 Dec, 2023</strong></p>
<p>Given an array of N integers where each value represents the number of chocolates in a packet. Each packet can have a variable number of chocolates. There are m students, the task is to distribute chocolate packets such that:</p>
<ul>
  <li>Each student gets <strong>one</strong> packet.</li>
  <li>The <strong>difference</strong> between the number of chocolates in the packet with <strong>maximum chocolates</strong> and the packet with <strong>minimum chocolates</strong> given to the students is <strong>minimum</strong>.</li>
</ul>
<h3>Examples:</h3>
<p><strong>Input :</strong> arr[] = {7, 3, 2, 4, 9, 12, 56} , m = 3<br>
<strong>Output:</strong> Minimum Difference is 2<br>
<strong>Explanation:</strong><br>
We have seven packets of chocolates and we need to pick three packets for 3 students<br>
If we pick 2, 3 and 4, we get the minimum difference between maximum and minimum packet sizes.</p>
<p><strong>Input :</strong> arr[] = {3, 4, 1, 9, 56, 7, 9, 12} , m = 5<br>
<strong>Output:</strong> Minimum Difference is 6</p>
<p><strong>Input :</strong> arr[] = {12, 4, 7, 9, 2, 23, 25, 41, 30, 40, 28, 42, 30, 44, 48, 43, 50} , m = 7<br>
<strong>Output:</strong> Minimum Difference is 10</p>


- Initially sort the given array. And declare a variable to store the minimum difference, and initialize it to INT_MAX. Let the variable be min_diff.
- Find the subarray of size m such that the difference between the last (maximum in case of sorted) and first (minimum in case of sorted) elements of the subarray is minimum.
- We will run a loop from 0 to (n-m), where n is the size of the given array and m is the size of the subarray.
- We will calculate the maximum difference with the subarray and store it in diff = arr[highest index] – arr[lowest index]
- Whenever we get a diff less than the min_diff, we will update the min_diff with diff.

```python
def ChocolateDistributionProblem(a,m):
    size = len(a)
    a.sort()

    min_diff = a[size-1] - a[0]
    # Edge cases
    if m == 0 or size == 0:
        return 0
    if size < m:
        return -1

    for i in range(size-m+1):
        min_diff = min(min_diff, a[i+m-1]-a[i])

    return min_diff
```

```python
a1, m1 = [7, 3, 2, 4, 9, 12, 56], 3
a2, m2 = [3, 4, 1, 9, 56, 7, 9, 12], 5
a3, m3 = [12, 4, 7, 9, 2, 23, 25, 41, 30, 40, 28, 42, 30, 44, 48, 43, 50], 7
```

```python
print(ChocolateDistributionProblem(a1, m1))
print(ChocolateDistributionProblem(a2, m2))
print(ChocolateDistributionProblem(a3, m3))
```
