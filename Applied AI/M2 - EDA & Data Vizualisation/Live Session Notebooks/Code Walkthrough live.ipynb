# Scatter Plot

```python
import seaborn as sns
sns.set(style='white')

# Load the example mpg dataset
mpg = sns.load_dataset('mpg')
print(type(mpg))
```

```python
# pretty print DataFrame
from tabulate import tabulate

print(tabulate(mpg, headers='keys',tablefmt='github'))
```

```python
# Plot mpg against horsepower

sns.scatterplot(x='horsepower', y='mpg', data=mpg)
```

```python
# Plot mpg against horsepower -- Adding grid
sns.set()

sns.scatterplot(x='horsepower', y='mpg', data=mpg)
```

```python
# Plot mpg against horsepower -- Adding grid
# Adding hue & title
sns.set()

sns.scatterplot(x='horsepower', y='mpg', data=mpg, hue='origin').set_title('ScatterPlot')
```

```python
# Plot mpg against horsepower -- Adding grid
# Adding hue & title
sns.set()

sns.scatterplot(x='horsepower', y='mpg', data=mpg, hue='origin',size='weight').set_title('ScatterPlot')
```

```python

sns.set()

sns.scatterplot(x='horsepower', y='mpg', data=mpg, hue='origin',size='weight', sizes=(10,200)).set_title('ScatterPlot')
```

```python

sns.set()

sns.scatterplot(x='horsepower', y='mpg', data=mpg, hue='origin',size='weight', sizes=(10,200), palette='muted').set_title('ScatterPlot')
```

```python
import matplotlib.pyplot as plt
sns.set()
plt.style.use('dark_background')
sns.scatterplot(x='horsepower', y='mpg', data=mpg, hue='origin',size='weight', sizes=(10,200), palette='muted').set_title('ScatterPlot')
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.set(style='ticks', color_codes=True)

g = sns.FacetGrid(tips, col='time', row='smoker')
print(type(g))
```

```python
import numpy as np

bins = np.arange(0,65,3)
#g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(plt.hist, 'total_bill', bins=bins)
```

```python
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```

```python
g = sns.FacetGrid(tips, col='time', hue='smoker')
(g.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')).add_legend()
```

```python
bins = np.arange(0,65,3)
g = sns.FacetGrid(tips, col='day',height=4)
(g.map(plt.hist, 'total_bill', bins=bins)).add_legend()
```

```python
bins = np.arange(0,65,3)
g = sns.FacetGrid(tips, col='day', row='smoker' ,height=4)
(g.map(plt.hist, 'total_bill', bins=bins)).add_legend()
```
