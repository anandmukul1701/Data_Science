```python
# One way to calculate the norm of a vector
w_norm = m.sqrt(np.dot(w,w))
print(w_norm)
```

```python
# Another way to get the norm of the vector
print(np.linalg.norm(w,2))  # 2 here justify the kind of norm we're looking for

w_hat = w/w_norm

print(w_hat)
print(np.linalg.norm(w_hat,2))
```

```python
from scipy import signal 
from scipy import misc

ascent = misc.ascent()
print(ascent.shape)

img_plot = plt.imshow(ascent, cmap = 'gray', vmin=0, vmax=255)
```

```python
k = 10
kernel = np.ones((k,k))
kernel /= np.sum(kernel)

print(kernel)

blurred = signal.convolve2d(ascent, kernel)

plt.imshow(blurred, cmap = 'gray')
plt.show()
```

```python
k = 20
kernel = np.ones((k,k))
kernel /= np.sum(kernel)

print(kernel)

blurred = signal.convolve2d(ascent, kernel)

plt.imshow(blurred, cmap = 'gray')
plt.show()
```
