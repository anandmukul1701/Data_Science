```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp

x = np.random.normal(loc = 20, scale = 5, size = 100)
sp.stats.probplot(x, dist='norm', plot=plt)
```

```python
x = np.random.exponential(scale=1.0, size=1000)

sns.set()
ax = sns.displot(x)
plt.show()
```

```python
x = np.random.exponential(scale=10, size=1000)

sns.set()
ax = sns.displot(x)
plt.show()
```

```python
kwargs = {'cumulative': True}
sns.displot(x, hist_kws=kwargs, kde_kws = kwargs)
```

```python
## QQ Plot
sp.stats.probplot(x, dist='norm', plot=plt)
plt.grid()
```
