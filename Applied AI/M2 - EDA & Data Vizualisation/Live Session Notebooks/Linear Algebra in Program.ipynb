```python
# One way to calculate the norm of a vector
w_norm = m.sqrt(np.dot(w,w))
print(w_norm)
```

```python
# Another way to get the norm of the vector
print(np.linalg.norm(w,2))  # 2 here justify the kind of norm we're looking for

w_hat = w/w_norm

print(w_hat)
print(np.linalg.norm(w_hat,2))
```
