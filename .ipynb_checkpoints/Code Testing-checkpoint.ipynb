```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 89]

# Plot with gridlines behind the bars
plt.figure(figsize=(10, 5))
plt.bar(categories, values, zorder=2)
plt.grid(True, zorder=1)
plt.title('Gridlines Behind Bars')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Plot with gridlines on top of the bars
plt.figure(figsize=(10, 5))
plt.bar(categories, values, zorder=1)
plt.grid(True, zorder=2)
plt.title('Gridlines On Top of Bars')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```
